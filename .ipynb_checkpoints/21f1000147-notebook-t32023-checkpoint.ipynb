```python
import numpy as np
import pandas as pd
import datetime
import seaborn as sns
sns.set_theme(style="whitegrid")

traindf = pd.read_csv("/kaggle/input/taxi-fare-guru-total-amount-prediction-challenge/train.csv")
testdf = pd.read_csv("/kaggle/input/taxi-fare-guru-total-amount-prediction-challenge/test.csv")
traindf['tpep_pickup_datetime'] = pd.to_datetime(traindf['tpep_pickup_datetime'], format='%Y-%m-%d %H:%M:%S')
traindf['tpep_dropoff_datetime'] = pd.to_datetime(traindf['tpep_dropoff_datetime'], format='%Y-%m-%d %H:%M:%S')

# Calculate the time difference
traindf['TripDuration'] = abs((traindf['tpep_dropoff_datetime'] - traindf['tpep_pickup_datetime']).dt.total_seconds())

traindf['PUDate']=traindf['tpep_pickup_datetime'].dt.date.map(datetime.datetime.toordinal)
traindf['PUTime']=traindf['tpep_pickup_datetime'].dt.hour
#traindf['PUTime'] = traindf['PUTime'].apply(lambda x: x.hour * 3600 + x.minute * 60 + x.second)

#traindf['DODate']=traindf['tpep_dropoff_datetime'].dt.date.map(datetime.datetime.toordinal)
#traindf['DOTime']=traindf['tpep_dropoff_datetime'].dt.time
#traindf['DOTime'] = traindf['DOTime'].apply(lambda x: x.hour * 3600 + x.minute * 60 + x.second)

traindf['PUDay'] = traindf['tpep_pickup_datetime'].dt.dayofweek

# traindf['PUDate']=pd.to_numeric(traindf.PUDate)
# traindf['PUTime']=pd.to_numeric(traindf.PUTime)
# traindf['PUTime'].corr(traindf['total_amount']),
traindf.head(20)
#traindf['PUDate'].corr(traindf['total_amount'])

```

```python
drop_columns = ['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'store_and_fwd_flag']
XNew = X.drop(columns=drop_columns)
```

```python
XNew.head()
```

```python
XNew['payment_type'].replace('unknown', 'Cash', inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(XNew, y, test_size=0.2, random_state=42)
```

```python
si = SimpleImputer(strategy="most_frequent")
ohe = OneHotEncoder(sparse_output=False)
le = LabelEncoder()
mms = MinMaxScaler()

std = StandardScaler()
tran_list = [('impute', si, ['passenger_count']), ('cat_encoder', ohe, ['payment_type']), ('scaler', std, ['trip_distance', 'tip_amount', 'extra', 'TripDuration', 'PUDate', 'PUTime', 'PUDay'])]
col_tran = ColumnTransformer(transformers=tran_list, remainder='passthrough', verbose_feature_names_out=False).set_output(transform="pandas")
tran_X = col_tran.fit_transform(X_train)
tran_X_test = col_tran.transform(X_test)
tran_X.head()
```

```python
tran_XNew = pd.DataFrame(tran_X)
tran_X_testNew = pd.DataFrame(tran_X_test)
```
